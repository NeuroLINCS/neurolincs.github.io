# Proteomics

```python
protein = pd.read_csv('./protein_abundances_JL.tsv', sep='\t')
protein.head()
```

```python
protein[protein.duplicated('Symbol', keep=False)]
```

```python
protein = protein.drop(protein.index[[1142]])
protein.loc[1141:1144]
```

```python
protein = protein.set_index('Symbol')
protein.head()
```

```python
protein.to_json('protein.json')
```

# Transcriptomics

```python
transcripts = pd.read_csv('./rna_abundances_JL.tsv', sep='\t')
transcripts.head()
```

```python
transcripts[transcripts.duplicated('symbol', keep=False)]
```

```python
transcripts = transcripts.groupby('symbol').sum(axis=1)
```

```python
transcripts = transcripts.reset_index().dropna(axis=0, subset=['symbol']).set_index('symbol')
```

```python
transcripts = transcripts[transcripts.sum(axis=1) != 0]
```

```python
transcripts.to_json('transcripts.json')
```
